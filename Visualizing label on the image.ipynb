```python
import pandas as pd
boxes = pd.read_csv('C:/Users/Admin/OneDrive - Iowa State University/Thesis documents/newdata/Labels_text/img_01_425391600_00018.txt', sep = ' ', header = None)

#1st box
x1 = (boxes[1][0] - boxes[3][0]*0.5)*frame.shape[1]
y1 = (boxes[2][0] - boxes[4][0]*0.5)*frame.shape[0]
w1 = boxes[3][0] * frame.shape[1]
h1 = boxes[4][0] * frame.shape[0]

#2nd box
x2 = (boxes[1][1] - boxes[3][1]*0.5)*frame.shape[1]
y2 = (boxes[2][1] - boxes[4][1]*0.5)*frame.shape[0]
w2 = boxes[3][1] * frame.shape[1]
h2 = boxes[4][1] * frame.shape[0]
```
